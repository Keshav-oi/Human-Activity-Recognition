```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dask.dataframe as dd
```

```python
classes = ["Laying", "Sitting", "Standing", "Walking", "Walking_downstairs", "Walking_upstairs"]
data = []
for i in range(len(classes)):
    data.append(dd.read_csv(classes[i] + "/*.csv"))
```

```python
column = ["a_x", "a_y", "a_z"]
laying_data = pd.DataFrame(data[0], columns=column)
sitting_data = pd.DataFrame(data[1], columns=column)
standing_data = pd.DataFrame(data[2], columns=column)
walking_data = pd.DataFrame(data[3], columns=column)
walking_downstairs_data = pd.DataFrame(data[4], columns=column)
walking_upstairs_data = pd.DataFrame(data[5], columns=column)
```

```python
laying_linear_accl = pd.DataFrame(((laying_data.iloc[:, 0]**2) + (laying_data.iloc[:, 0]**2) + (laying_data.iloc[:, 0]**2))**0.5)
sitting_linear_accl = pd.DataFrame(((sitting_data.iloc[:, 0]**2) + (sitting_data.iloc[:, 0]**2) + (sitting_data.iloc[:, 0]**2))**0.5)
standing_linear_accl = pd.DataFrame(((standing_data.iloc[:, 0]**2) + (standing_data.iloc[:, 0]**2) + (standing_data.iloc[:, 0]**2))**0.5)
walking_linear_accl = pd.DataFrame(((walking_data.iloc[:, 0]**2) + (walking_data.iloc[:, 0]**2) + (walking_data.iloc[:, 0]**2))**0.5)
walking_downstairs_linear_accl = pd.DataFrame(((walking_downstairs_data.iloc[:, 0]**2) + (walking_downstairs_data.iloc[:, 0]**2) + 
                                  (walking_downstairs_data.iloc[:, 0]**2))**0.5)
walking_upstairs_linear_accl = pd.DataFrame(((walking_upstairs_data.iloc[:, 0]**2) + (walking_upstairs_data.iloc[:, 0]**2) + 
                                (walking_upstairs_data.iloc[:, 0]**2))**0.5)
```

```python
time = np.linspace(0,10,500)
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
plt.plot(time, laying_data.iloc[:500, 0])
plt.plot(time, laying_data.iloc[:500, 1])
plt.plot(time, laying_data.iloc[:500, 2])
plt.plot(time, laying_linear_accl[:500].values)
plt.legend(["a_x", "a_y", "a_z", "linear_acc"]), plt.xlabel("Time->"), plt.ylabel('Acceleration'), plt.title("Laying")

plt.subplot(2, 2, 2)
plt.plot(time, sitting_data.iloc[:500, 0])
plt.plot(time, sitting_data.iloc[:500, 1])
plt.plot(time, sitting_data.iloc[:500, 2])
plt.plot(time, sitting_linear_accl[:500].values)
plt.legend(["a_x", "a_y", "a_z", "linear_acc"]), plt.xlabel("Time->"), plt.ylabel('Acceleration'), plt.title("Sitting")

plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
plt.plot(time, standing_data.iloc[:500, 0])
plt.plot(time, standing_data.iloc[:500, 1])
plt.plot(time, standing_data.iloc[:500, 2])
plt.plot(time, standing_linear_accl[:500].values)
plt.legend(["a_x", "a_y", "a_z", "linear_acc"]), plt.xlabel("Time->"), plt.ylabel('Acceleration'), plt.title("Standing")

plt.subplot(2, 2, 2)
plt.plot(time, walking_data.iloc[:500, 0])
plt.plot(time, walking_data.iloc[:500, 1])
plt.plot(time, walking_data.iloc[:500, 2])
plt.plot(time, walking_linear_accl[:500].values)
plt.legend(["a_x", "a_y", "a_z", "linear_acc"]), plt.xlabel("Time->"), plt.ylabel('Acceleration'), plt.title("Walking")

plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
plt.plot(time, walking_downstairs_data.iloc[:500, 0])
plt.plot(time, walking_downstairs_data.iloc[:500, 1])
plt.plot(time, walking_downstairs_data.iloc[:500, 2])
plt.plot(time, walking_downstairs_linear_accl[:500].values)
plt.legend(["a_x", "a_y", "a_z", "linear_acc"]), plt.xlabel("Time->"), plt.ylabel('Acceleration'), plt.title("Walking_Downstairs")

plt.subplot(2, 2, 2)
plt.plot(time, walking_upstairs_data.iloc[:500, 0])
plt.plot(time, walking_upstairs_data.iloc[:500, 1])
plt.plot(time, walking_upstairs_data.iloc[:500, 2])
plt.plot(time, walking_upstairs_linear_accl[:500].values)
plt.legend(["a_x", "a_y", "a_z", "linear_acc"]), plt.xlabel("Time->"), plt.ylabel('Acceleration'), plt.title("Walking_upstairs")
plt.show()
```
