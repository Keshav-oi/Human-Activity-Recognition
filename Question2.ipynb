```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dask.dataframe as dd
```

```python
classes = ["Laying", "Sitting", "Standing", "Walking", "Walking_downstairs", "Walking_upstairs"]
data = []
for i in range(len(classes)):
    data.append(dd.read_csv(classes[i] + "/*.csv"))
```

```python
laying_data = data[0]
sitting_data = data[1]
standing_data = data[2]
walking_data = data[3]
walking_downstairs_data = data[4]
walking_upstairs_data = data[5]
```

```python
laying_linear_accl = pd.DataFrame(((laying_data.iloc[:, 0]**2) + (laying_data.iloc[:, 0]**2) + (laying_data.iloc[:, 0]**2))**0.5)
sitting_linear_accl = pd.DataFrame(((sitting_data.iloc[:, 0]**2) + (sitting_data.iloc[:, 0]**2) + (sitting_data.iloc[:, 0]**2))**0.5)
standing_linear_accl = pd.DataFrame(((standing_data.iloc[:, 0]**2) + (standing_data.iloc[:, 0]**2) + (standing_data.iloc[:, 0]**2))**0.5)
walking_linear_accl = pd.DataFrame(((walking_data.iloc[:, 0]**2) + (walking_data.iloc[:, 0]**2) + (walking_data.iloc[:, 0]**2))**0.5)
walking_downstairs_linear_accl = pd.DataFrame(((walking_downstairs_data.iloc[:, 0]**2) + (walking_downstairs_data.iloc[:, 0]**2) + 
                                  (walking_downstairs_data.iloc[:, 0]**2))**0.5)
walking_upstairs_linear_accl = pd.DataFrame(((walking_upstairs_data.iloc[:, 0]**2) + (walking_upstairs_data.iloc[:, 0]**2) + 
                                (walking_upstairs_data.iloc[:, 0]**2))**0.5)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2, 2, 1)
plt.hist(laying_linear_accl, bins = 50), plt.title("laying_linear_accl"), plt.ylabel("No. of Values")

plt.subplot(2, 2, 2)
plt.hist(sitting_linear_accl, bins = 50), plt.title("sitting_linear_accl"), plt.ylabel("No. of Values")

plt.figure(figsize=(15,10))
plt.subplot(2, 2, 1)
plt.hist(standing_linear_accl, bins = 50), plt.title("standing_linear_accl"), plt.ylabel("No. of Values")

plt.subplot(2, 2, 2)
plt.hist(walking_linear_accl, bins = 50), plt.title("walking_linear_accl"), plt.ylabel("No. of Values")

plt.figure(figsize=(15,10))
plt.subplot(2, 2, 1)
plt.hist(walking_downstairs_linear_accl, bins = 50), plt.title("walking_downstairs_linear_accl"), plt.ylabel("No. of Values")

plt.subplot(2, 2, 2)
plt.hist(walking_upstairs_linear_accl, bins = 50), plt.title("walking_upstairs_linear_accl"), plt.ylabel("No. of Values")
plt.show()
```

As seen from the graphs, linear acceleration for Laying can be easily distinguished from sitting and standing activity

### Linear Acceleration: $ (a_x^2 + a_y^2 + a_z^2)^(0.5) $

- Laying: Linear acceleration is always less than 0.5 $ m/s^2 $. Which can be assumed as zero, since we don't any activity during laying
- Sitting: Linear acceleration lies between 1.4 to 1.8, which is greater than laying activity.
- Standing: Linear acceleration lies between 1.6 to 1.8, which is nearly equal to sitting acitvity.

And for dynamic activities, linear acceleration varies from 0.5 to 3 $ m/s^2 $.


```python
plt.hist(laying_linear_accl, bins = 50)
plt.hist(sitting_linear_accl, bins = 50)
plt.hist(standing_linear_accl, bins = 50)
plt.hist(walking_linear_accl, bins = 50)
plt.hist(walking_downstairs_linear_accl, bins = 50)
plt.hist(walking_upstairs_linear_accl, bins = 50)
plt.legend(["Laying", "Sitting", "Standing", "Walking", "Walking_downstairs", "Walking_upstairs"])
plt.show()
```

### As seen from this graph, static activities can be easily distinguish from dynamic activities.

- Linear acceleration for static activities (standing, sitting) is striclty between 1.6 to 1.7.
- Linear acceleration for dynamic activities (Walking, Walking_downstairs, Walking_upstairs) varies from 0.5 to 3.5.
- While Linear acceleration for Laying is between 0 to 0.5.

### Hence, for a large subset of data, we can distinguish without Machine Learning model, only when we are considering Linear acceleration in the case.
